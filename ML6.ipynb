# CONSIDERING THE WHOLE CONTAINER



```
company=soup.find_all('div',class_='company-content-wrapper')
```

```
len(company)
```

```
name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
how_old=[]
no_of_employee=[]

for i in company:

  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  reviews.append(i.find('a' , class_='review-count').text.strip())
  ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
  how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
  no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())

df=pd.DataFrame({'name':name,
   'rating':rating,
   'reviews':reviews,
   'company_type':ctype,
   'Head_Quarters':hq,
   'Company_Age':how_old,
   'No_of_Employee':no_of_employee,
   })

```

# creating dataframe for all the pages

```
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

final = pd.DataFrame()

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"

# Add a User-Agent header to the request to avoid being blocked
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
webpage = requests.get(url, headers=headers).text
soup = BeautifulSoup(webpage, "lxml")

table = soup.find("table", class_="wikitable")
rows = table.find_all("tr")[1:]   # skip header row

name = []
industry = []
revenue = []
employees = []
hq = []

for row in rows:
    cols = row.find_all("td")

    try:
        name.append(cols[0].text.strip())
    except:
        name.append(np.nan)

    try:
        industry.append(cols[1].text.strip())
    except:
        industry.append(np.nan)

    try:
        revenue.append(cols[2].text.strip())
    except:
        revenue.append(np.nan)

    try:
        employees.append(cols[3].text.strip())
    except:
        employees.append(np.nan)

    try:
        hq.append(cols[4].text.strip())
    except:
        hq.append(np.nan)

df = pd.DataFrame({
    "Company_Name": name,
    "Industry": industry,
    "Revenue": revenue,
    "Employees": employees,
    "Headquarters": hq
})

final = pd.concat([final, df], ignore_index=True)

print(final.head())
```
